```python
import pandas as pd
from datasets import load_dataset
```

```python
nonce_words = open('nonce_words.txt').readlines()
```

```python
nonce_words = [word.strip() for word in nonce_words]
```

```python
nonce_words
```

```python
affixes = load_dataset('joey234/affixal_negation')
affixes = affixes['train']['affix']
affixes = list(set(affixes))
affixes.remove('-')
affixes.remove('less')
affixes.remove('free')
# affixes.remove('ef')
# affixes.remove('ir')


prefixes = affixes
```

```python
suffixes = ['less','free']
```

```python
non_neg_affixes = open('non-neg-affix.txt').readlines()
```

```python
non_neg_affixes = [x.strip() for x in non_neg_affixes]
```

```python
non_neg_affixes
```

```python
nonce_words
```

```python
words = []
affixes = []
for nonce in nonce_words:
    for prefix in prefixes:
        word = prefix + '-' + nonce
        # word = prefix  + nonce
        
        words.append(word)
        affixes.append(prefix)
    for suffix in suffixes:
        word = nonce+ '-' + suffix
        # word = nonce + suffix
        
        words.append(word)
        affixes.append(suffix)
```

```python
import tiktoken
from transformers import AutoTokenizer
```

```python
def has_negative_affix(tokens, affix):
    # if len(tokens)==1:
    #     return False
    for token in tokens:
        if token == affix:
            return True
    return False
```

```python
hf_model = 'google/flan-t5-xxl'
tokenizer = AutoTokenizer.from_pretrained(hf_model)
```

```python
for i,word in enumerate(words):
    token_integers = tokenizer(word).input_ids
    num_tokens = len(token_integers)
    token_bytes = [tokenizer.decode(token, skip_special_tokens = True, clean_up_tokenization_spaces = True).strip('#') for token in token_integers]
    token_bytes = list(filter(None, token_bytes))
    # for 
    if has_negative_affix(token_bytes, affixes[i]) == False:
        print(word, token_bytes)
        # only_negative_affix_dict[key].append(word)
```

```python
non_neg_words = []
# non_neg_affixes = []
non_neg_affixes_to_write = []
for word in nonce_words:
    # print(word)
    for affix in non_neg_affixes:
        # print(affix)
        if affix.startswith('-'):
            tmp_word =  word + affix
        elif affix.endswith('-'):
            tmp_word = affix+word
        non_neg_affixes_to_write.append(affix.strip('-'))
        non_neg_words.append(tmp_word)
```

```python
labels_non_neg = [1]* len(non_neg_words)
```

```python
from datasets import Dataset
```

```python
ds = Dataset.from_dict({'word': words + non_neg_words , 'affix': affixes + non_neg_affixes_to_write, 'label': labels + labels_non_neg })
```

```python
shuffled_dataset = ds.shuffle(seed=42)
```
